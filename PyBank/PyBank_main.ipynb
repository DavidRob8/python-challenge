```python
import os
import csv

# Locate csv file
budget_data_csv = os.path.join("budget_data.csv")

# Open csv file
with open(budget_data_csv) as csv_file:
    csv_reader = csv.reader(csv_file,delimiter=',')
    
    # Skips the header row
    header = next(csv_reader)
    
    # Define variables
    month_total = 0
    net_total_amount = 0
    previous_pl = 0
    changes = []
    greatest_increase = ["", 0]
    greatest_decrease = ["", 0]
    
    # Initiate For loop to loop through rows
    for row in csv_reader:
        
        # Total number of months
        month_total = month_total + 1
        
        # Net total of profit and loss through entire period
        net_total_amount = net_total_amount + int(row[1])
        
        # Calculates the change in profit and loss over the entire period
        if month_total > 1:
            change = int(row[1]) - previous_pl
            changes.append(change)
            
            # The greatest increase in profits (date and amount) over the entire period
            if change > greatest_increase[1]:
                greatest_increase[0] = row[0]
                greatest_increase[1] = change
                
            # The greatest decrease in profits (date and amount) over the entire period
            if change < greatest_decrease[1]:
                greatest_decrease[0] = row[0]
                greatest_decrease[1] = change
                
        # set previous profit and loss to previous values
        previous_pl = int(row[1])
        
    # Calculates the avergae of profit and loss changes
    avg = sum(changes)/len(changes)
    
    # Print findings
    #print("Financial Analysis")
    #print("----------------------------")
    #print("Total Months: "+ str(month_total))
    #print("Total: " + str(net_total_amount))
    #print("Average Change: $" + "{:.2f}".format(avg))
    #print("Greatest Increase in Profits: "+ greatest_increase[0] + " ($" + str(greatest_increase[1]) + ")")
    #print("Greatest Decrease in Profits: "+ greatest_decrease[0] + " ($" + str(greatest_decrease[1]) + ")")
                
    # Open a text file for writing
    output_file = "financial_analysis.txt"
    with open(output_file, 'w') as output_file:
        print("Financial Analysis", file=output_file)
        print("----------------------------", file=output_file)
        print("Total Months: " + str(month_total), file=output_file)
        print("Total: " + str(net_total_amount), file=output_file)
        print("Average Change: $" + "{:.2f}".format(avg), file=output_file)
        print("Greatest Increase in Profits: " + greatest_increase[0] + " ($" + str(greatest_increase[1]) + ")", file=output_file)
        print("Greatest Decrease in Profits: " + greatest_decrease[0] + " ($" + str(greatest_decrease[1]) + ")", file=output_file)
        
        # Print findings
        print("Financial Analysis")
        print("")
        print("----------------------------")
        print("")
        print("Total Months: " + str(month_total))
        print("")
        print("Total: " + str(net_total_amount))
        print("")
        print("Average Change: $" + "{:.2f}".format(avg))
        print("")
        print("Greatest Increase in Profits: " + greatest_increase[0] + " ($" + str(greatest_increase[1]) + ")")
        print("")
        print("Greatest Decrease in Profits: " + greatest_decrease[0] + " ($" + str(greatest_decrease[1]) + ")")
```
