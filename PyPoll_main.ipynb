```python
import csv
import os

# Open csv

with open ("election_data.csv", "r") as file:
    csv_reader = csv.reader(file)
    
    # Skip header row
    next(csv_reader)
    
    # Define variables
    total_votes = 0
    dict_candidates = {}
    most_votes = 0
    winner = ""
    
    #Loop through rows
    for row in csv_reader:
        
        # total number of votes
        total_votes += 1
        
        # Extract candidate name from row
        candidate_name = row[2]
        
        # If candidate name is not in the dict, add candidate name
        if candidate_name not in dict_candidates:
            dict_candidates[candidate_name] = 0
            
        # Total number of votes each candidate won
        dict_candidates[candidate_name] += 1
        
    # Open a text file for writing
    output_file = "election_analysis.txt"
    with open(output_file, 'w') as output_file:
        print("Election Results", file=output_file)
        print("", file=output_file)
        print("----------------------------", file=output_file)
        print("", file=output_file)
        print(f"Total Votes: {total_votes}", file=output_file)
        print("", file=output_file)
        print("----------------------------", file=output_file)
        print("", file=output_file)
        for candidate_name in dict_candidates:
            votes = dict_candidates[candidate_name]
            percentage = (votes / total_votes) * 100
            print(f"{candidate_name}: {percentage:.3f}% ({votes})", file=output_file)
            print("", file=output_file)
            
            if votes > most_votes:
                most_votes = votes
                winner = candidate_name
                
        print("----------------------------", file=output_file)
        print("", file=output_file)
        print(f"Winner: {winner}", file=output_file)
        print("", file=output_file)
        print("----------------------------", file=output_file)
    
        # Print findings
        print("Election Results")
        print("")
        print("---------------------------------------------")
        print("")

        # Print the total number of votes
        print(f"Total Votes: {total_votes}")
        print("")
        print("---------------------------------------------")
        print("")

        # Print the percentage of votes for each candidate
        for candidate_name in dict_candidates:
            votes = dict_candidates[candidate_name]
            percentage = (votes / total_votes) * 100
            print(f"{candidate_name}: {percentage:.3f}% ({votes})")
            print("")
    
            # Keep track of the winner
            if votes > most_votes:
                most_votes = votes
                winner = candidate_name
          
        print("---------------------------------------------")
        print("")
    
        # Print the winner of the election based on popular vote
        print(f"Winner: {winner}")
        print("")
        print("---------------------------------------------")
```
